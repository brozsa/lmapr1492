```python
from pymatgen import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator


with MPRester("6D8NXvBA170Om5w3") as m:
    structure = m.get_structure_by_material_id("mp-2530")
```

```python
c = XRDCalculator()
```

```python
c.show_plot(structure)
```

```python
c.get_pattern(structure)
```

```python
pattern = c.get_pattern(structure)
two_theta = pattern.x
intensity = pattern.y
hkls = pattern.hkls
```

```python
#PREMIER PIC
print("pic de diffraction 2=" , two_theta[0])
print("intensit :" , intensity[0])
print("indice hkl associs et sa multiplicit :" , hkls[0])
```

```python
#DEUXIME PIC
print("pic de diffraction 2=" , two_theta[1])
print("intensit :" , intensity[1])
print("indice hkl associs et sa multiplicit :" , hkls[1])
```
